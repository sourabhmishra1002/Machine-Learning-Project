```python
#Null values in the dataset
df.isnull().sum()
```

The isnull.sum() command is use to see the number of the null values in the different columns. Here 4 columns name,host_name,last_review and reviews_per_month have 16,21,10052 and 10052 respectively have null numbers.

```python
#Handling the null values

# Fill missing values in 'reviews_per_month' with 0
df.fillna({'reviews_per_month': 0}, inplace=True)

# Replace missing 'last_review' with 'NaT' in datetime format
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Drop rows where essential fields like 'name' or 'host_name' are missing
df.dropna(subset=['name', 'host_name'], inplace=True)
```

Null values in any column can cause serious distortion in the presentation or processing of the data. In order to counter the problems related to the null values there are several methods to handle the null value efficiently.

```python
#Checking the null value handling
df.isnull().sum()
```

Here we have seen that now null values are handled and now only one column out of 4 colums have null values i.e last_review.

```python
#Numerical exploration of the data
df.describe()
```

The above table displays all the measures such as central tendencies, quartiles and range of the values in the columns of the data.

```python
#Outliers detection

# Boxplot for price
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Price Boxplot')
plt.show()
```

Outliers are extreme values that differ from most other data points in a dataset. They can have a big impact on your statistical analyses and skew the results of any hypothesis tests.
The box plot above show the outliers in the price column.

```python
#Compare prices across room types

plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df,color="orange")
plt.title('Price by Room Type')
plt.ylabel('Price')
plt.xlabel('Room Type')
plt.show()
```

The above boxplot shows the comparison between interquartile range of price and type of rooms.

```python
# Distribution of price
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True,color="orange")
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

The above graph shows the price distribution.


```python
# Bar plot for room types
plt.figure(figsize=(6, 6))
sns.countplot(x='room_type', data=df, order=df['room_type'].value_counts().index,color="orange")
plt.title('Room Type Distribution')
plt.ylabel('Count')
plt.show()
```

The above barplot depicts the number of rooms of different types. The quantity of entire home is the highest while the shared rooms are least in the count.

```python
# Average Availability by Room Type
plt.figure(figsize=(8, 6))
avg_availability = df.groupby('room_type')['availability_365'].mean().sort_values()
avg_availability.plot(kind='bar', color='orange')
plt.title('Average Availability by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Availability (days)')
plt.show()
```

The above barplot shows that average availabilty of shared room is greatest followed by entire home and private room.

```python
#Geographical Analysis

plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='price', size='price', sizes=(20, 200), data=df, palette='viridis')
plt.title('Airbnb Listings by Location')
plt.show()
```

The scatter plot above indicates the density and spread of the properties across the nation.

```python
# Average Price by Neighbourhood Group

plt.figure(figsize=(8, 6))
avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values()
avg_price.plot(kind='bar', color='orange')
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()
```

The above graph depicts the relation between Neighbourhood Group and Average price. The average price of Manhattan is highest and Bronx

```python
# Top 10 Neighbourhoods by Number of Listings

plt.figure(figsize=(10, 6))
top_neighbourhoods = df['neighbourhood'].value_counts().head(10)
sns.barplot(x=top_neighbourhoods.values, y=top_neighbourhoods.index, palette='pastel')
plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.show()
```

The above graph shows Neighbourhood vs Number of listings of the dataset.
Willimsburg has the highest number while Midtown has least number of the listing.

```python
# Price Distribution by Room Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df, palette='pastel')
plt.ylim(0, 1000)  # Focus on reasonable price range
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

The above boxplot depicts the range in the interquartile range of price for different type of the rooms.

```python
# Reviews vs. Availability
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='availability_365', hue='neighbourhood_group',
                data=df, alpha=0.6, palette='viridis')
plt.title('Reviews vs. Availability')
plt.xlabel('Number of Reviews')
plt.ylabel('Availability (days)')
plt.legend(title='Neighbourhood Group', loc='upper right')
plt.show()
```

The above graph show the availability and number of reviews with different color dots represents different areas.

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df[['price', 'minimum_nights', 'number_of_reviews',
                           'reviews_per_month', 'availability_365']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

The above correlation matrix shows the relation and dependency of one column over another column.
